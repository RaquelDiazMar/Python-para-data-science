```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("base_datos_2008.csv",nrows=1000000)
data = np.unique(df.DayOfWeek,return_counts=True)
labs = ['Lun','Mar','Mie','Jue','Vie','Sab','Dom']
data
```

```python
plt.pie(x=data[1],
       labels=data[0],
       radius=1.5,
       colors=['Red','Green','Orange','Blue','Gray','Pink','Black'],
       startangle=90)
plt.show()
```

```python
plt.pie(x=data[1],
       labels=labs,
       radius=1.5,
       colors=sns.color_palette('hls',7),
       explode=(0,0,0,.2,0,0,.1),
       startangle=90,
       autopct='%1.1f%%',
       labeldistance=1.5)
plt.show()
```

```python
plt.pie(x=data[1],
#        labels=labs,
       radius=.7,
       colors=sns.color_palette('hls',7),
       explode=(0,0,0,.2,0,0,.1),
       startangle=90,
       autopct='%1.1f%%')
plt.legend(loc='upper left', labels=labs)
plt.show()
```

```python
plt=sns.barplot(x=labs,y=data[1])
plt.set(xlabel='Da de la Semana',ylabel='Nmero de Vuelos')
```
