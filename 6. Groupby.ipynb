```python
# Para realizar el siguiente tipo de análisis, que es un poco más complejo, vamos a crear una 
#nueva base de datos, que la llamaremos 'dfATLHOU'
dfATLHOU = df[df.Origin.isin(["ATL","HOU"])]
```

```python
dfATLHOU.groupby(by = ["DayOfWeek","Origin"])["ArrDelay"].mean()
# se comparan varias columnas agrupadas
```

```python
dfATLHOU.groupby(by = ["Origin","DayOfWeek"])["ArrDelay"].mean()
# el orden de la agrupación es importante
```

```python
mygroupby=dfATLHOU.groupby(by = ["Origin","DayOfWeek"])["ArrDelay"]
# los objetos 'groupby' pueden guardarse
```

```python
mygroupby.mean()
```

```python
mygroupby.describe()
```
