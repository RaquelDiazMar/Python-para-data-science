```python
import pandas as pd
import numpy as np
from joblib import Parallel, delayed

df=pd.read_csv("base_datos_2008.csv",nrows=100000)
```

```python
df_sub = df[['CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay']]
df_sub.head(10)
```

```python
# definir la función que se va a aplicar sobre cada una de las filas de esta base de datos,
# y va a servir para calcular dentro de cada una de estas filas en las que se pueda buscar cuál 
# es el retraso máximo en caso de que se pueda. Lo primero es definir una condición, que va a 
# buscar si tenemos "any" en la fila. Vamos a definir los nombres de columna y vamos a devolver 
# el nombre de entre todos los valores de la fila, su posición que haya tomado un valor máximo. 
# En caso contrario, vamos a devolver ninguno.

def retraso_max(fila):
    if not np.isnan(fila).any():
        names = ['CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay']
        return names[fila.index(max(fila))]
    else:
        return 'None'
```

```python
results=[]
for fila in df_sub.values.tolist():
    results.append(max(fila))
```

```python
results
```

```python
result = Parallel(n_jobs = 2, backend = 'multiprocessing')(map(delayed(max),df_sub.values.tolist()))
# n_jobs = nº de procesadores a usar
```
