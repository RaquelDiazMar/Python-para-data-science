```python
import pandas as pd
df = pd.read_csv("C:/Users/34677/Documents/GitHub/Python para data science y big data esencial/base_datos_2008.csv",nrows=1e6)
```

```python
dfduplicate = df.append(df)
```

```python
dfduplicate=dfduplicate.sample(frac=1)
# para reordenar el dataframe
```

```python
dfclean = dfduplicate.drop_duplicates()
# con esta función se eliminan los datos duplicados
```

```python
len(dfclean) == len(df)
# comprobación de que se han eliminado los duplicados
```

```python
len(dfclean)
# df contaba con 1e6 filas, al tener dfclean 4 menos, significa que df tenía 2 filas duplicadas
```

```python
dfclean.drop_duplicates(subset="DayofMonth")
#  buscar duplicados sobre la columna que le especifiquemos, por ejemplo, 'DayofMonth'. 
# Si ejecutamos, podemos ver que ha mantenido una sola fila para cada distinto valor de 
# 'DayofMonth', así que nos hemos quedado con una base de datos con tan solo 31 días. 
```

```python
# gestión de datos faltantes
dfclean.dropna()
#  obtenemos un "data frame" vacío. ¿Por qué ha ocurrido eso? Básicamente porque ninguna de las filas de nuestro "data frame" no tiene ningún dato faltante. Esto es debido a que, si el vuelo está cancelado, no tiene hora de llegada. Y si tiene hora de llegada, el vuelo no está cancelado. Así que por fuerza alguna de estas dos columnas va a tener un dato faltante.
```

```python
dfclean.head()
```

```python
dfclean.dropna(thresh=25)
# con el parámetro threshold especificamos cuántas entradas de todas las columnas que tenemos queremos que estén llenas de información relevante. Por ejemplo, podríamos decirle "quiero 25 entradas llenas de las 29 que hay". 
```

```python
dfclean.dropna(subset=["CancellationCode"])
# Si lo que nos interesa es filtrar por una columna en concreto, podemos usar la misma función y usar el parámetro 'subset', y dentro de una lista ponerle el nombre de columna por el que queremos filtrar. Así, nos va a devolver un "data frame" donde no tenga NaN en esta columna. 
```
