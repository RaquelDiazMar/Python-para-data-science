```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# https://matplotlib.org/
# Recommneded to visit: 'Tutorials' & 'Examples'

df = pd.read_csv("base_datos_2008.csv",nrows=100000)
```

```python
data = np.unique(df.Cancelled, return_counts=True)
data
# numpy.unique(arr, return_index, return_inverse, return_counts)
# arr: The input array. Will be flattened if not 1-D array
# return_counts: If True, returns the number of times the element in unique array appears in the original array
```

```python
plt.pie(x=data[1],
       labels=data[0])
plt.show()
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pie.html
```

```python
plt.pie(x=data[1],
       labels=data[0],
       colors=['Red','Green'],
       shadow=True,
       startangle=90,
       radius=2)
plt.show()
```
